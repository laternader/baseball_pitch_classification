### get_batter()

---
Create a function to extract the batters name since we don't have it.

````python
def get_batter(des):
    try:
        name = ' '.join(des.split(' ',2)[:2])
        return name
    except:
        return np.nan # goal is to return previous name 
````

````python
mlb2020['des'].head()
````

````python
get_batter(mlb2020['des'][1]) 
````

````python
# Create column called 'batter'
mlb2020['batter_name'] = mlb2020['des'].map(get_batter)
````

### Pitch Type Data Dictionary

---
Fastballs
- FA = fastball
- FF = four-seam fastball
- FT = two-seam fastball
- FC = fastball (cutter)
- FS / SI / SF = fastball (sinker, split-fingered)

Off-Speed
- SL = slider
- CH = changeup
- CB / CU = curveball
- KC = knuckle-curve
- KN = knuckleball
- EP = eephus

Other
- UN / XX = unidentified
- PO / FO = pitch out
