### get_batter()

---
Create a function to extract the batters name from `des` column since we don't have it.

````python
def get_batter(des):
    try:
        name = ' '.join(des.split(' ',2)[:2])
        return name
    except:
        return np.nan # goal is to return previous name 
````

````python
mlb2020['des'].head()
````

````python
# to check how it works on NaNs
get_batter(mlb2020['des'][1]) 
````

````python
# Create column called 'batter'
mlb2020['batter_name'] = mlb2020['des'].map(get_batter)
````

### Create a function that filters all dataframes by name

````python
def pitcher_stats(name ,df):
    '''
    name: name of pitcher as a string
    df: dataframe of the season you want to check
    '''
    try:
        return df[df['player_name']==name]
    except:
        return 'Did not play this season or you misspelled his name.'
````

````python
Julio_Urias_19 = pitcher_stats('Julio Urias', mlb2019)
````

````python
Julio_Urias_19.head()
````

````python
def batter_stats(name ,df):
    '''
    name: name of pitcher as a string
    df: dataframe of the season you want to check
    '''
    try:
        return df[df['batter_name']==name]
    except:
        return 'Did not play this season or you misspelled his name.'
````

### Pitch Type Data Dictionary

---
Fastballs
- FA = fastball
- FF = four-seam fastball
- FT = two-seam fastball
- FC = fastball (cutter)
- FS / SI / SF = fastball (sinker, split-fingered)

Off-Speed
- SL = slider
- CH = changeup
- CB / CU = curveball
- KC = knuckle-curve
- KN = knuckleball
- EP = eephus

Other
- UN / XX = unidentified
- PO / FO = pitch out
