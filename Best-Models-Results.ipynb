# Metrics

```python
def get_metrics(model, X_train, X_test, y_train, y_test):
    
    # Predictions
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)
    train_probs = model.predict_proba(X_train)
    test_probs = model.predict_proba(X_test)
    
    # f1 score
    print('Train F1 Score:{}'.format(f1_score(y_train, train_preds, average='weighted')))
    print('Test F1 Score:{}'.format(f1_score(y_test, test_preds, average='weighted')))
    
    # accuracy
    print(f'Train Accuracy Score:{accuracy_score(y_train, train_preds)}')
    print(f'Test Accuracy Score:{accuracy_score(y_test, test_preds)}')
    
    # classification
    print(f'Train Classification Report:\n{classification_report(y_train, train_preds)}')
    print(f'Test Classification Report:\n{classification_report(y_test, test_preds)}')
    
    # roc
    print('Train ROC OVO Score:{}'.format(roc_auc_score(y_train, train_probs, average='weighted', multi_class='ovo')))
    print('Train ROC OVR Score:{}'.format(roc_auc_score(y_train, train_probs, average='weighted', multi_class='ovr')))
    print('Test ROC OVO Score:{}'.format(roc_auc_score(y_test, test_probs, average='weighted', multi_class='ovo')))
    print('Test ROC OVR Score:{}'.format(roc_auc_score(y_test, test_probs, average='weighted', multi_class='ovr')))
```

```python
get_metrics(type_rfmodel_name, type_train, type_test, y1_train, y1_test)
```

```python
get_metrics(strike_rfmodel_name, strike_train, strike_test, y2_train, y2_test)
```
