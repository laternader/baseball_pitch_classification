```python
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBRFClassifier, XGBClassifier
from sklearn.metrics import plot_confusion_matrix, f1_score, roc_curve

import pickle

pd.set_option('display.max_columns', None)
```

```python
mlb2020 = pd.read_csv('data/cleaned2020.csv')
```

```python
X = mlb2020.drop(columns=['index','events', 'bb_type', 'on_3b', 'on_2b', 'on_1b', 'pitch_type', 'pitch_name', 'game_date', 'inning_topbot', 'home_team', 'away_team'])
y = mlb2020['pitch_name']
```

```python
X
```

## LabelEncode the Y variable

```python
label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(y)
label_encoded_y = label_encoder.transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, label_encoded_y, random_state=66, stratify=y)
```

# rfModel

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

# XGBoost

```python
xgb = XGBClassifier()

xgb.fit(X_train, y_train)
```

# Get Dummies

```python
dummy_train = pd.get_dummies(X_train, drop_first=False, sparse=True)
```

```python
dummy_test = pd.get_dummies(X_test, drop_first=False)
```

```python
dummy_train
```

```python
dummy_train.shape, dummy_test.shape
```

```python
[c for c in dummy_train.columns if c not in dummy_test.columns]
```

```python
[c for c in dummy_test.columns if c not in dummy_train.columns]
```

```python
dummy_train.drop(columns=[c for c in dummy_train.columns if c not in dummy_test.columns],inplace=True)
```

```python
xgb = XGBClassifier()

xgb.fit(dummy_train, y_train)
```

```python
xgb.score(dummy_train, y_train) # Mean Accuracy Score of Train
```

```python
xgb.score(dummy_test, y_test) # Mean Accuracy Score of Test
```

```python
# Run this to full send
preds = xgb.predict(dummy_test)
```

```python
f1_score(y_test, preds, average=None)
```
